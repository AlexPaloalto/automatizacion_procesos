Se ajustan los pagares afectados, ya que se van a menor precio, primero lo categorizamos y los identificamos. Todo lo que este fuera de la ciudad de México y este entre enero del 2020 y julio del 2022, esta afectado

```python
control_df["fecha_disposicion"] = pd.to_datetime(control_df["fecha_disposicion"])

# Extraer año y mes
control_df["Año"] = control_df["fecha_disposicion"].dt.year
control_df["Mes"] = control_df["fecha_disposicion"].dt.month

# Definir condición de fecha
condicion_fecha = (
    (control_df["Año"] == 2020) | 
    (control_df["Año"] == 2021) | 
    ((control_df["Año"] == 2022) & (control_df["Mes"] < 8))
)

# Definir condición de sucursal
condicion_sucursal = ~control_df["estado"].isin(["CDMX", "CIUDAD DE MÉXICO", "DISTRITO FEDERAL"])

# Asignar el estado de afectación
control_df["Estado"] = np.where(condicion_fecha & condicion_sucursal, "Afectado", "No Afectado")
```

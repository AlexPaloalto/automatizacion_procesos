```python
import pandas as pd
import numpy as np

# 📌 Cargar el historial de asignaciones
try:
    municipios_previos = pd.read_excel("municipios_1.0.xlsx")
except FileNotFoundError:
    municipios_previos = pd.DataFrame(columns=["municipio", "M_mayor"])  # En caso de que el archivo no exista

# 📌 Generar la nueva tabla dinámica
municipios = control_df.pivot_table(values="Precio", index="municipio", aggfunc={"Precio": ["count"]})
municipios = municipios.reset_index()
print(len(municipios))

# 📌 Generar aleatoriedad para la nueva asignación
municipios["Solicitud JM"] = np.random.rand(len(municipios))
municipios["Solicitud GB"] = 1 - municipios["Solicitud JM"]

# 📌 Determinar el coordinador asignado en esta versión (sin alternar aún)
municipios["M_mayor"] = np.where(municipios["Solicitud GB"] > municipios["Solicitud JM"], "GB", "JM")

# 🔥 **Alternar la asignación si el municipio ya tiene historial**
for index, row in municipios.iterrows():
    municipio = row["municipio"]
    
    # 📌 Buscar si ya tiene asignación previa
    asignacion_previa = municipios_previos.loc[municipios_previos["municipio"] == municipio, "M_mayor"]
    
    if not asignacion_previa.empty:
        # 📌 Si el municipio ya tenía un coordinador, alternarlo
        nuevo_asignado = "JM" if asignacion_previa.values[0] == "GB" else "GB"
        municipios.at[index, "M_mayor"] = nuevo_asignado

# 📌 Calcular porcentajes y reglas de división (manteniendo la lógica)
municipios["Porcentaje"] = np.where(
    (municipios["Solicitud GB"] > 0) & (municipios["Solicitud JM"] == 0), 
    1,
    np.where(
        (municipios["Solicitud JM"] > 0) & (municipios["Solicitud GB"] == 0),
        1,
        np.where(municipios["M_mayor"] == "GB", 
                 municipios["Solicitud JM"] / municipios["Solicitud GB"],
                 municipios["Solicitud GB"] / municipios["Solicitud JM"])
    )
)

municipios["Regla_Div"] = np.select(
    [
        (municipios["Porcentaje"] >= 0.0) & (municipios["Porcentaje"] <= 0.20),
        (municipios["Porcentaje"] > 0.20) & (municipios["Porcentaje"] <= 0.80),
        (municipios["Porcentaje"] > 0.80) & (municipios["Porcentaje"] < 1),
        (municipios["Porcentaje"] == 1)
    ],
    ["UNA-UNA", "UNA-UNA", "UNA-UNA", "UNA-UNA"],
    default="UNA-UNA"
)

# 💾 Exportar el nuevo DataFrame para que puedas consultar la asignación en futuros cálculos
municipios.to_excel("municipios_2.0.xlsx", index=False)

print("✅ Nueva asignación generada con alternancia! 🚀", len(municipios))
```
